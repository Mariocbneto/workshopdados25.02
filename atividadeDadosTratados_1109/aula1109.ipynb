```python
import pandas as pd
import unidecode 
data = r'dados_nao_tratados.csv'
df = pd.read_csv(data)
df.head(11)
```

```python
df.columns = (
    df.columns.str.strip()
              .str.upper()               
              .str.replace(' ', '_')     
              .str.replace('[^0-9A-Z_]', '', regex=True)  
)

```

```python


df['NOME'] = (
    df['NOME']
    .str.replace(r"\(.*?\)", "", regex=True)  
    .str.strip()
    .str.upper()                              
    .apply(unidecode.unidecode)               
)

```

```python
df['ATIVO'] = df['ATIVO'].str.upper().map({
    'SIM': True, 'TRUE': True, 'YES': True,
    'NÃO': False, 'NAO': False, 'FALSE': False, 'E': False
})
```

```python
df['ID'] = pd.to_numeric(df['ID'], errors='coerce')
df['IDADE'] = pd.to_numeric(df['IDADE'], errors='coerce')
df['DATA_INSCRICAO'] = pd.to_datetime(df['DATA_INSCRICAO'], errors='coerce', dayfirst=True)
df['NOTA'] = pd.to_numeric(df['NOTA'], errors='coerce')
df['DATA_INSCRICAO'] = df['DATA_INSCRICAO'].fillna('DESCONHECIDO')
df['IDADE'] = df['IDADE'].fillna('DESCONHECIDO')
df['NOTA'] = df['NOTA'].fillna('DESCONHECIDO')

```
