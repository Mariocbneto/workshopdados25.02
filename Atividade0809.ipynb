```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


temperaturas = np.random.randint(20, 36, size=7)  
dias = ["Seg", "Ter", "Qua", "Qui", "Sex", "Sáb", "Dom"]


df = pd.DataFrame({"Dia": dias, "Temperatura": temperaturas})
media = df["Temperatura"].mean()
print(df)
print("Média da semana:", media)


plt.plot(df["Dia"], df["Temperatura"], marker='o')
plt.title("Temperatura longo da semana")
plt.xlabel("Dia")
plt.ylabel("Temperatura (°C)")
plt.show()
```
