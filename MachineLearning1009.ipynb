```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import joblib

# 3. Carregar e explorar os dados
df = pd.read_csv('diabetes.csv')
print("Dimensões:", df.shape)
print(df.head())
print(df.info())
print(df.describe())
print("\nContagem de classes:")
print(df['Outcome'].value_counts())

# 4. Tratamento dos dados
cols_with_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in cols_with_zeros:
    df[col] = df[col].replace(0, np.nan)
    df[col].fillna(df[col].median(), inplace=True)

# 5. Separar base de dados em arrays
X = df.drop('Outcome', axis=1).values
y = df['Outcome'].values

# 6. Pré-processamento
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 7. Dividir base em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# 8. Definir modelos e treinar
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42)
}

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"\n=== {name} ===")
    print(f"Acurácia: {acc*100:.2f}%")
    print("Matriz de Confusão:")
    print(confusion_matrix(y_test, y_pred))
    print("\nRelatório de Classificação:")
    print(classification_report(y_test, y_pred))

# 9. Escolher melhor modelo
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"\nMelhor modelo: {best_model_name} com acurácia {results[best_model_name]*100:.2f}%")

# 10. Salvar solução
joblib.dump(best_model, 'melhor_modelo_diabetes.pkl')
joblib.dump(scaler, 'scaler_diabetes.pkl')
print("\nModelo e scaler salvos em /mnt/data/")
```
